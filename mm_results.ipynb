```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## 2x2
syn_2x2_16 = [157.31680458106527, 5973.600000000003, 0.25508736]
opt_2x2_16 = [163.56332517697547, 5730.840000000002, 0.199121472]
syn_2x2_32 = [130.87035328451861, 7258.150000000005, 0.794395008]
opt_2x2_32 = [130.87035328451861, 7258.150000000004, 0.772161984]
```

```python
## 4x4
syn_4x4_16 = [90.04944615088152,10722.009999999991,2.04069888]
opt_4x4_16 = [92.56936453908332,10419.709999999992,1.64691648]
syn_4x4_32 = [71.13109104424007, 13675.549999999983,6.355160064]
opt_4x4_32 = [71.4223984927018,13618.209999999981,6.247828224]
```

```python
## 8x8
syn_8x8_16 = [51.768114051367526,18933.909999999963,16.32559104]
opt_8x8_16 = [52.301501419201344, 18736.909999999963, 13.493703168]
syn_8x8_32 = [39.06336977034264, 25216.42999999994,50.841280512]
opt_8x8_32 = [39.12291127667067,25177.46999999994,50.2647552]
```

```python
cols = ["max_frequency", "timing_delay", "area"]
unit = ["2x2","4x4","8x8"]
syn_16 = pd.DataFrame([syn_2x2_16,syn_4x4_16,syn_8x8_16],columns=cols)
opt_16 = pd.DataFrame([opt_2x2_16,opt_4x4_16,opt_8x8_16],columns=cols)
syn_32 = pd.DataFrame([syn_2x2_32,syn_4x4_32,syn_8x8_32],columns=cols)
opt_32 = pd.DataFrame([opt_2x2_32,opt_4x4_32,opt_8x8_32],columns=cols)
```

```python
syn_16
```

```python
plt.plot(syn_16["area"],syn_16["timing_delay"], color="blue",marker='o',label='syn_16bit')
plt.plot(opt_16["area"],opt_16["timing_delay"], color="orange", marker='o',linestyle= "--", label='opt_16bit')
plt.plot(syn_32["area"],syn_32["timing_delay"], color="green",marker='o',label='syn_32bit')
plt.plot(opt_32["area"],opt_32["timing_delay"], color="red", marker='o',linestyle= "--", label='opt_32bit')
plt.xlabel("Area (mm^2)")
plt.ylabel("Max timing delay (ps)")
plt.title("Area vs Timing Delay")
plt.legend()
plt.show()
```

```python
plt.plot(syn_16["area"],syn_16["max_frequency"], color="blue",marker='o',label='syn_16bit')
plt.plot(opt_16["area"],opt_16["max_frequency"], color="orange", marker='o',linestyle= "--", label='opt_16bit')
plt.plot(syn_32["area"],syn_32["max_frequency"], color="green",marker='o',label='syn_32bit')
plt.plot(opt_32["area"],opt_32["max_frequency"], color="red", marker='o',linestyle= "--", label='opt_32bit')
plt.xlabel("Area (mm^2)")
plt.ylabel("Max frequency (MHz)")
plt.title("Area vs Max Frequency")
plt.legend()
plt.show()
```
